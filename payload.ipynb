```python
def bin_pixels(image, bin_factor):

    height, width = image.shape[:2]

    new_height = height // bin_factor
    new_width = width // bin_factor

    reshaped_image = image[:new_height * bin_factor, :new_width * bin_factor].reshape(new_height, bin_factor, new_width, bin_factor)

    binned_image = reshaped_image.sum(axis=(1, 3)) // (bin_factor ** 2)

    return binned_image.astype(np.uint8)
```
